```python
# Install required packages (run these in your terminal if not already installed)
# pip install chromadb sentence-transformers

import chromadb
from sentence_transformers import SentenceTransformer
```

```python
# Initialize Chroma client (this creates a local persistent vector database)
client = chromadb.Client()
collection = client.create_collection(name="sentence_embeddings")

# Initialize a sentence transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Example sentences to store
sentences = [
    "The sun rises in the east.",
    "I love playing football with my friends.",
    "Artificial intelligence is transforming industries.",
    "The cat is sleeping on the couch.",
    "Python is a popular programming language.",
    "He went to the store to buy some milk.",
    "Machine learning models learn from data.",
    "The weather today is very pleasant.",
    "She enjoys reading books in her free time.",
    "New York City is known as the city that never sleeps."
]

# Convert sentences into embeddings
embeddings = model.encode(sentences)

# Add sentences and embeddings to ChromaDB collection
collection.add(
    documents=sentences,
    embeddings=embeddings.tolist(),
    ids=[f"id{i}" for i in range(len(sentences))]
)

```

```python
# Input query sentence
query_sentence = "AI is changing how companies work."

# Convert query to embedding
query_embedding = model.encode([query_sentence])

# Perform semantic search in ChromaDB
results = collection.query(
    query_embeddings=query_embedding.tolist(),
    n_results=1  # retrieve the most similar sentence
)

# Display results
print("Query:", query_sentence)
print("\nMost similar sentence found:")
print(results["documents"][0][0])
```

This notebook demonstrates how to use ChromaDB and Sentence-Transformers for semantic search.

1. **Setup**: The first two cells install the necessary libraries (`chromadb` and `sentence-transformers`) and import them.
2. **ChromaDB Initialization**: The third cell initializes a ChromaDB client and creates a collection named "sentence_embeddings". It also loads a pre-trained sentence transformer model (`all-MiniLM-L6-v2`) which will be used to convert text into numerical representations (embeddings).
3. **Data Preparation and Loading**: The fourth cell defines a list of example sentences and converts them into embeddings using the initialized model. These sentences and their corresponding embeddings are then added to the ChromaDB collection.
4. **Semantic Search**: The fifth cell takes a query sentence, converts it into an embedding, and then uses ChromaDB to find the most similar sentence in the collection based on the embeddings. The most similar sentence is then printed.

In essence, the notebook sets up a small vector database of sentences and their embeddings, and then shows how to query this database with a new sentence to find the most semantically similar sentence.

The code is performing **semantic search**.

Semantic search is a type of similarity search that focuses on the meaning or context of the text rather than just keyword matching. In this notebook, we are using sentence embeddings to capture the semantic meaning of the sentences and find similar ones based on that meaning.
