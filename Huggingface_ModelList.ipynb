```python
from huggingface_hub import HfApi

api = HfApi()
models_generator = api.list_models(author="facebook") # Store the generator

models_generator = sorted(list(models_generator), key=lambda x:x.id)
for model in models_generator:
  print(model.id)

print("Done")
```
